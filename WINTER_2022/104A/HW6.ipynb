<b>1.)</b> Construct orthogonal polynomials of degree $0$, $1$, and $2$ on the interval $(0,1)$ with respect to the weight function:<br>
&emsp;<b>a.)</b> $w(x)=\log\frac{1}{x}$<br>
&emsp;<b>b.)</b> $w(x)=\frac{1}{\sqrt{x}}$<br>
<b>Proof:</b><br>
&emsp;<b>a.)</b> Letting $\{g_{1},g_{2},...,g_{n}\}$ be an orthonormal system in an inner-product space $E$, we can construct orthogonal polynomials $p_{0}(x),p_{1}(x),...,p_{n}(x)$ defined as:
$$p_{n}(x)=(x-a_{n})p_{n-1}(x)-b_{n}p_{n-2}(x)\quad(n\geq 2)\quad p_{0}(x)=1,\quad p_{1}(x)=x-a_{1}$$
&emsp; Where our unknowns are given by:
$$a_{n}=\frac{\langle xp_{n-1}\rangle, p_{n-1}\rangle}{\langle p_{n-1},p_{n-1}\rangle}$$
$$b_{n}=\frac{\langle xp_{n-1}\rangle, p_{n-2}\rangle}{\langle p_{n-2},p_{n-2}\rangle}$$
$$\langle f,g\rangle=\int_{a}^{b}f(x)g(x)w(x)dx$$
&emsp; Therefore, we have that $p_{0}=1$, $p_{1}=x-a_{1}$, and $p_{2}=(x-a_{2})p_{1}(x)-b_{2}p_{0}(x)=(x-a_{2})(x-a_{1})-b_{2}$. Now solving for $a_{1},a_{2},b_{2}$:
$$a_{1}=\frac{\langle x,1\rangle}{\langle 1,1\rangle}=\frac{\int_{0}^{1}x\log\frac{1}{x}dx}{\int_{0}^{1}\log\frac{1}{x}dx}=\frac{\frac{1}{4}}{1}=\frac{1}{4}$$
$$a_{2}=\frac{\langle x(x-\frac{1}{4}),(x-\frac{1}{4})\rangle}{\langle(x-\frac{1}{4}),(x-\frac{1}{4})\rangle}=\frac{\int_{0}^{1}x\left(x-\frac{1}{4}\right)^{2}\log\frac{1}{x}dx}{\int_{0}^{1}\left(x-\frac{1}{4}\right)^{2}\log\frac{1}{x}dx}=\frac{\frac{13}{576}}{\frac{7}{144}}=\frac{13}{28}$$
$$b_{2}=\frac{\langle x(x-\frac{1}{4}),1\rangle}{\langle1,1\rangle}=\frac{\int_{0}^{1}\left(x^{2}-\frac{1}{4}x\right)\log\frac{1}{x}dx}{\int_{0}^{1}\log\frac{1}{x}dx}=\frac{\frac{7}{144}}{1}=\frac{7}{144}$$
&emsp; Altogether, for degree $0\Rightarrow p_{0}(x)=1$, for degree $1\Rightarrow p_{1}(x)=x-\frac{1}{4}$, for degree $2\Rightarrow p_{2}(x)=(x-\frac{13}{28})(x-\frac{1}{4})-\frac{7}{144}$<br>
&emsp;<b>b.)</b> Similarly, we find our $a_{1},a_{2},b_{2}$:
$$a_{1}=\frac{\langle x,1\rangle}{\langle 1,1\rangle}=\frac{\int_{0}^{1}x\frac{1}{\sqrt{x}}dx}{\int_{0}^{1}\frac{1}{\sqrt{x}}dx}=\frac{\frac{2}{3}}{2}=\frac{1}{3}$$
$$a_{2}=\frac{\langle x\left(x-\frac{1}{3}\right),\left(x-\frac{1}{3}\right)\rangle}{\langle\left(x-\frac{1}{3}\right),\left(x-\frac{1}{3}\right)\rangle}=\frac{\int_{0}^{1}x\left(x-\frac{1}{3}\right)^{2}\frac{1}{\sqrt{x}}dx}{\int_{0}^{1}{\left(x-\frac{1}{3}\right)^{2}\frac{1}{\sqrt{x}}dx}}=\frac{\frac{88}{945}}{\frac{8}{45}}=\frac{11}{21}$$
$$b_{2}=\frac{\langle x(x-\frac{1}{3}),1\rangle}{\langle1,1\rangle}=\frac{\int_{0}^{1}\left(x^{2}-\frac{1}{3}x\right)\frac{1}{\sqrt{x}}dx}{\int_{0}^{1}\frac{1}{\sqrt{x}}dx}=\frac{\frac{8}{45}}{2}=\frac{4}{45}$$
&emsp; So we conclude with $p_{0}(x)=1$, $p_{1}(x)=x-\frac{1}{3}$, $p_{2}(x)=\left(x-\frac{11}{21}\right)\left(x-\frac{1}{3}\right)-\frac{4}{45}$<br>

<b>2.)</b> In each of the following, find the least square approximation of degrees $0$, $1$, and $2$ for the function $f(x)=\sin(x)$ on the interval $(a,b)$ with respect to the weight function $w(x)=1$.<br>
&emsp;<b>a.)</b> $(a,b)$ = $(-\pi,\pi)$<br>
&emsp;<b>b.)</b> $(a,b)$ = $(-\frac{\pi}{2},\frac{\pi}{2})$<br>
<b>Proof:</b><br>
&emsp;<b>a.)</b> Similar to <b>Homework 5, Problem 6</b>, we use:
$$E(a_{0},a_{1},...,a_{n})=\sum_{j=1}^{m}(y_{i}-\sum_{k=0}^{n}a_{k}x_{i}^{k})^{2}$$
&emsp; Then for degree $0$, we have the following:
$$E(a_{0})=\int_{-\pi}^{\pi}(\sin(x)-a_{0})^{2}\Rightarrow \frac{\partial E}{a_{0}}=2\int_{-\pi}^{\pi}(\sin(x)-a_{0})dx=0$$
$$\Rightarrow (-\cos(x)-a_{0}x)\bigg|_{-\pi}^{\pi}=0$$
$$\Rightarrow (-\cos(\pi)-a_{0}\pi)-(\cos(-\pi)+a_{0}\pi)=0$$
$$\Rightarrow 2a_{0}\pi=0\Rightarrow a_{0}=0$$
&emsp; So for degree $0$, $p_{0}(x)=0$<br>
&emsp; For degree $1$, we have:
$$E(a_{0},a_{1})=\int_{-\pi}^{\pi}(\sin(x)-(a_{0}+a_{1}))^{2}$$
&emsp; Solving for $a_{0}$:
$$\frac{\partial E}{\partial a_{0}}=-2\int_{-\pi}^{\pi}(\sin(x)-a_{0}-a_{1}x)dx=0$$
$$\Rightarrow (-\cos(x)-a_{0}x-\frac{1}{2}a_{1}x^{2})\bigg|_{-\pi}^{\pi}=0$$
$$\Rightarrow (1-a_{0}\pi-\frac{1}{2}a_{1}\pi^{2})-(1+a_{0}\pi-\frac{1}{2}a_{1}\pi^{2})=0$$
$$\Rightarrow -2a_{0}\pi=0\Rightarrow a_{0}=0$$
&emsp; Then for $a_{1}$:
$$\frac{\partial E}{\partial a_{1}}=-2\int_{-\pi}^{\pi}(\sin(x)-a_{0}-a_{1}x)(x)dx=0$$
$$\Rightarrow \int_{-\pi}^{\pi}x\sin(x)dx-\int_{-\pi}^{\pi}a_{0}xdx-\int_{-\pi}^{\pi}a_{1}x^{2}dx=0$$
$$\Rightarrow (\sin(x)-x\cos(x))\bigg|_{\pi}^{0}-\frac{1}{2}a_{0}x^{2}\bigg|_{\pi}^{0}-\frac{1}{3}a_{1}x^{3}\bigg|_{\pi}^{0}=0$$
$$\Rightarrow 2\pi-0-\frac{2\pi^{3}}{9}a_{1}=0\Rightarrow a_{1}=\frac{9}{\pi^{2}}$$
&emsp; Therefore, for degree $1$, $p_{0}(x)=\frac{9}{\pi^{2}}x$<br>
&emsp; Then, for degree $2$:
$$E(a_{0},a_{1},a_{2})=\int_{-\pi}^{\pi}(\sin(x)-a_{0}-a_{1}x-a_{2}x^{2})dx=0$$
&emsp; For one equation,
$$\frac{\partial E}{\partial a_{0}}=-2\int_{-\pi}^{\pi}(\sin(x)-a_{0}-a_{1}x-a_{2}x^{2})dx=0$$
$$\Rightarrow (-\cos(x)-a_{0}x-\frac{1}{2}a_{1}x^{2}-\frac{1}{3}x^{3})\bigg|_{-\pi}^{\pi}=0$$
$$\Rightarrow -2\pi a_{0}-\frac{2}{3}\pi^{3}a_{2}=0\Rightarrow a_{0}=\frac{1}{3}a_{2}$$
&emsp; Then the next,
$$\frac{\partial E}{\partial a_{1}}=-2\int_{-\pi}^{\pi}(\sin(x)-a_{0}-a_{1}x-a_{2}x^{2})(x)dx=0$$
$$\Rightarrow \int_{-\pi}^{\pi}(x\sin(x)-a_{0}x-a_{1}x^{2}-a_{2}x^{3})dx=0$$
$$\Rightarrow (\sin(x)-x\cos(x)-\frac{1}{2}a_{0}x^{2}-\frac{1}{3}a_{1}x^{3}-\frac{1}{4}a_{2}x^{4})\bigg|_{-\pi}^{\pi}=0$$
$$\Rightarrow 2\pi-\frac{2}{3}a_{1}\pi^{3}=0\Rightarrow a_{1}=\frac{3}{\pi^{3}}$$
&emsp; Finally,
$$\frac{\partial E}{\partial a_{2}}=-2\int_{-\pi}^{\pi}(\sin(x)-a_{0}-a_{1}x-a_{2}x^{2})(x^{2})dx=0$$
$$\Rightarrow \int_{-\pi}^{\pi}(x^{2}\sin(x)-a_{0}x^{2}-a_{1}x^{3}-a_{2}x^{4})dx=0$$
$$\Rightarrow (2x\sin(x)+(2-x^{2})\cos(x)-\frac{1}{3}a_{0}x^{3}-\frac{1}{4}a_{1}x^{4}-\frac{1}{5}a_{2}x^{5})\bigg|_{-\pi}^{\pi}=0$$
$$\Rightarrow -\frac{2}{3}\pi^{3}a_{0}-\frac{2}{5}\pi^{5}a_{2}=0\Rightarrow a_{0}=\frac{3}{5}\pi^{2}a_{2}$$
&emsp; Therefore, the system of equations gives us that $a_{0}=0$, $a_{1}=\frac{3}{\pi^{3}}$, $a_{2}=0$. So for degree $2$, $p_{2}=\frac{3}{\pi^{3}}x$<br>
&emsp; <b>b.)</b> Again, for degree $0$,
$$E(a_{0})=\int_{\frac{-\pi}{2}}^{\frac{\pi}{2}}(\sin(x)-a_{0})^{2}\Rightarrow \frac{\partial E}{a_{0}}=2\int_{\frac{-\pi}{2}}^{\frac{\pi}{2}}(\sin(x)-a_{0})dx=0$$
$$\Rightarrow (-\cos(x)-a_{0}x)\bigg|_{\frac{-\pi}{2}}^{\frac{\pi}{2}}=0\Rightarrow -\pi a_{0}=0\Rightarrow a_{0}=0$$
&emsp; So, $p_{0}(x)=0$.<br>
&emsp; For degree $1$:
$$E(a_{0},a_{1})=\int_{-\pi}^{\pi}(\sin(x)-(a_{0}+a_{1}))^{2}$$
$$\Rightarrow \frac{\partial E}{\partial a_{0}}=-2\int_{\frac{-\pi}{2}}^{\frac{\pi}{2}}(\sin(x)-a_{0}-a_{1}x)dx=0$$
$$\Rightarrow (x\sin(x)-a_{0}x-\frac{1}{2}a_{1}x^{2})\bigg|_{-\frac{\pi}{2}}^{\frac{\pi}{2}}=0$$
$$\Rightarrow \pi a_{0}=0\Rightarrow a_{0}=0$$
$$\frac{\partial E}{\partial a_{1}}=-2\int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}(\sin(x)-a_{0}-a_{1}x)(x)dx=0$$
$$\Rightarrow \int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}(x\sin(x)-a_{0}x-a_{1}x^{2}-a_{2}x^{3})dx=0$$
$$\Rightarrow (\sin(x)-x\cos(x)-\frac{1}{2}a_{0}x^{2}-\frac{1}{3}a_{1}x^{3})\bigg|_{-\frac{\pi}{2}}^{\frac{\pi}{2}}=0$$
$$\Rightarrow 2-\frac{\pi^{3}}{12}a_{1}=0\Rightarrow a_{1}=\frac{6}{\pi^{3}}$$
&emsp; Therefore, $p_{1}=\frac{6}{\pi^{3}}x$.<br>
&emsp; Lastly, for degree $2$:
$$E(a_{0},a_{1},a_{2})=\int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}(\sin(x)-a_{0}-a_{1}x-a_{2}x^{2})dx=0$$
$$\frac{\partial E}{\partial a_{0}}=-2\int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}(\sin(x)-a_{0}-a_{1}x-a_{2}x^{2})dx=0$$
$$\Rightarrow (-\cos(x)-a_{0}x-\frac{1}{2}a_{1}x^{2}-\frac{1}{3}x^{3})\bigg|_{-\frac{\pi}{2}}^{\frac{\pi}{2}}=0$$
$$\Rightarrow -\frac{1}{2}\pi a_{0}-\frac{1}{12}\pi^{3}a_{1}\Rightarrow a_{0}=\frac{1}{6}a_{1}$$
$$\frac{\partial E}{\partial a_{1}}=-2\int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}(\sin(x)-a_{0}-a_{1}x-a_{2}x^{2})(x)dx=0$$
$$\Rightarrow \int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}(x\sin(x)-a_{0}x-a_{1}x^{2}-a_{2}x^{3})dx=0$$
$$\Rightarrow (\sin(x)-x\cos(x)-\frac{1}{2}a_{0}x^{2}-\frac{1}{3}a_{1}x^{3}-\frac{1}{4}a_{2}x^{4})\bigg|_{-\frac{\pi}{2}}^{\frac{\pi}{2}}=0$$
$$\Rightarrow 2-\frac{1}{12}a_{1}\pi^{3}=0\Rightarrow a_{1}=\frac{24}{\pi^{3}}$$
$$\frac{\partial E}{\partial a_{2}}=-2\int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}(\sin(x)-a_{0}-a_{1}x-a_{2}x^{2})(x^{2})dx=0$$
$$\Rightarrow \int_{-\frac{\pi}{2}}^{\frac{\pi}{2}}(x^{2}\sin(x)-a_{0}x^{2}-a_{1}x^{3}-a_{2}x^{4})dx=0$$
$$\Rightarrow (2x\sin(x)+(2-x^{2})\cos(x)-\frac{1}{3}a_{0}x^{3}-\frac{1}{4}a_{1}x^{4}-\frac{1}{5}a_{2}x^{5})\bigg|_{-\frac{\pi}{2}}^{\frac{\pi}{2}}=0$$
$$\Rightarrow -\frac{1}{12}\pi^{3}a_{0}-\frac{1}{80}\pi^{5}a_{2}=0\Rightarrow a_{0}=\frac{1}{20}\pi^{2}a_{2}$$
&emsp; We conclude, $p_{2}=\frac{24}{\pi^{3}}x$<br>
<b>3.)</b> Let $x_{0},x_{1},...,x_{n}$ be distinct points in $[a,b]$ and we interpolate a function $f(x)$ by a polynomial $p_{n}(x)$ by Lagrange interpolation method. Let $\{l_{j}(x)^{(n)}\}_{j=0}^{n}$ be the Lagrange polynomials used in Lagrange interpolation. Prove $\sum_{j=0}^{n}l_{j}(x)^{(n)}=1$ for all $x$.<br>
<b>Proof:</b><br>
&emsp; We recall that $l_{j}(x)$ is defined as:
$$l_{j}(x)=\prod_{j=0, j\neq i}^{n}\frac{x-x_{j}}{x_{i}-x_{j}}\quad(0\leq i\leq n)$$
&emsp; And the Lagrange polynomial is:
$$p(x)=\sum_{j=0}^{n}y_{j}l_{j}(x)$$
&emsp; Suppose we interpolate the function $y(x)=1$, then $1=\sum_{j=1}^{n}(1)l_{j}(x)=y(x)$, thereby showing what we want and interpolating our function perfectly. Furthermore, since $p(x)$ is unique, it must always be the case that $\sum_{j=0}^{n}l_{j}(x)=1$ for all $x$.<br>
<b>4.)</b> Let $x_{0},x_{1},...,x_{n}$ be distinct points in $[a,b]$. Show that if $f$ and its first derivatives are defined respectively at points $x_{0},x_{1},...,x_{n}$, then there <b>exists a unique</b> polynomial $q$ of degree at most $2n+1$ such that
$$q(x_{j})=f(x_{j}),\quad\textrm{and}\quad q'(x_{j})=f'(x_{j})$$
&emsp; for $j=0,1,...,n$.<br>
<b>Proof:</b><br>
&emsp; Assume there exists a different polynomial $t(x)$ where $\textrm{Deg}(t(x))\leq 2n+1$ such that
$$t(x_{j})=f(x_{j}),\quad\textrm{and}\quad t'(x_{j})=f'(x_{j})$$
&emsp; for $j=0,1,...,n$. Then the polynomial $r(x)=t(x)-q(x)=0$ and $r'(x)=t'(x)-q'(x)=0$ is a polynomial of degree $2n+1$ with roots of multiplicity $2$ of $x_{0},x_{1},...,x_{n}$. To prove uniqueness, we let $s(x)$ be some polynomial, and $r(x)=s(x)(x-x_{0})^2(x-x_{1})^{2}...(x-x_{n})^{2}$. Then $s(x)$ must be $0$ since $r(x)$ should have $n$ zeros by the Fundamental Theorem of Algebra. If it is the case that $s(x)\neq 0$, then $\textrm{Deg}(r(x))>2n+1$ and $r(x)=0$. If $r(x)=0$, it is implied that $q(x)=t(x)$ therefore, $q(x)$ is unique.
